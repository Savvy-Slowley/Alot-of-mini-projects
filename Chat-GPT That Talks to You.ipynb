```python
# Import necessary libraries
# openai to interact with OpenAI API
import openai
# gTTS to convert text to speech
from gtts import gTTS
# IPython.display to play mp3 file
from IPython.display import Audio
# import the SpeechRecognition library
import speech_recognition as sr

# Function to generate text responses from OpenAI API
def askGPT(text):
    # Replace "INSERT YOUR API SECRET KEY HERE" with your actual API key
    openai.api_key = "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"
    # Use the OpenAI API to generate text responses
    response = openai.Completion.create(
        engine = "text-davinci-003",
        prompt = text,
        temperature = 0.6,
        max_tokens = 150,
    )
    # Print the text response
    print(response.choices[0].text)
    # convert the response text to speech using Jravis voice
    tts = gTTS(response.choices[0].text, lang='en')
    # save the response in mp3 format
    tts.save("response.mp3")
    # play the mp3 file
    return Audio("response.mp3", autoplay=True)

# Main function to take user input and call the askGPT function
def main():
    while True:
        # Initialize a new Recognizer object
        r = sr.Recognizer()
        # Use the microphone as the audio source
        with sr.Microphone() as source:
            print("Speak:")
            # Listen for audio input from the microphone
            audio = r.listen(source)
            try:
                # Use the Recognizer object to convert speech to text
                myQn = r.recognize_google(audio)
                print("You said: " + myQn)
                audio = askGPT(myQn)
                # play the audio automatically
                display(audio)
                input("Press Enter to continue...")
                print('\n')
            except sr.UnknownValueError:
                print("Could not understand audio")
            except sr.RequestError as e:
                print("Could not request results; {0}".format(e))

# Run the main function
main()
```
